### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = x * 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error checks
if len(x) != len(y):
    raise ValueError("x and y must have the same length.")
if len(categories) != len(values):
    raise ValueError("categories and values must have the same length.")
if len(labels) != len(sizes):
    raise ValueError("labels and sizes must have the same length.")
if any(size < 0 for size in sizes):
    raise ValueError("Pie sizes must be non-negative.")

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Line Plot
axes[0, 0].plot(x, y, marker='o', color='blue', label='y = 2x')
axes[0, 0].set_title("Line Chart", fontsize=14)
axes[0, 0].set_xlabel("x values")
axes[0, 0].set_ylabel("y values")
axes[0, 0].legend()
axes[0, 0].grid(True, linestyle='--', alpha=0.6)

# 2. Bar Chart
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart", fontsize=14)
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# 3. Scatter Plot
axes[1, 0].scatter(x, y, color='orange', edgecolors='black', s=80)
for i, j in zip(x, y):
    axes[1, 0].annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(-10, 10), ha='center', fontsize=8)
axes[1, 0].set_title("Scatter Plot", fontsize=14)
axes[1, 0].set_xlabel("x values")
axes[1, 0].set_ylabel("y values")
axes[1, 0].grid(True, linestyle='--', alpha=0.5)

# 4. Pie Chart
axes[1, 1].pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=['gold', 'lightblue', 'lightgreen', 'coral'],
    explode=(0.1, 0, 0, 0),
    shadow=True,
    textprops=dict(fontsize=10)
)
axes[1, 1].set_title("Pie Chart", fontsize=14)

# Final layout adjustment
plt.tight_layout()

# Display all subplots
plt.show()
```
