### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'lightblue', 'lightgreen', 'coral']  # Custom colors

# Error handling
if len(sizes) != len(labels):
    raise ValueError("The 'sizes' and 'labels' lists must have the same length.")
if any(size < 0 for size in sizes):
    raise ValueError("The 'sizes' values must all be non-negative.")
if sum(sizes) == 0:
    raise ValueError("The total size must be greater than zero.")

# Create figure
plt.figure(figsize=(8, 6), dpi=100)

# Create pie chart
wedges, texts, autotexts = plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    explode=(0.1, 0, 0, 0),
    shadow=True,
    textprops=dict(color="black", fontsize=10)
)

# Improve text appearance
for text in autotexts:
    text.set_fontweight('bold')

# Title
plt.title("Programming Language Popularity", fontsize=16, fontweight='bold')

# Optional: Equal aspect ratio to ensure circle shape
plt.axis('equal')

# Optional: Save the chart
# plt.savefig("programming_language_pie_chart.png", dpi=300)

# Show plot
plt.tight_layout()
plt.show()
```
