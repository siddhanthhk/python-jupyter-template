### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'lightblue', 'lightgreen', 'coral']  # Optional: Customize colors

# Error handling
if len(sizes) != len(labels):
    raise ValueError("The 'sizes' and 'labels' lists must have the same length.")
if any(size < 0 for size in sizes):
    raise ValueError("The 'sizes' values must all be non-negative.")
if sum(sizes) == 0:
    raise ValueError("The total size must be greater than zero.")

# Create the pie chart
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',  # Display percentages with 1 decimal place
    colors=colors,
    startangle=90,  # Start the pie chart at 90 degrees
    explode=(0.1, 0, 0, 0),  # Slightly "explode" the first slice
    shadow=True  # Add shadow for better visual effect
)

# Add a title
plt.title("Programming Language Popularity", fontsize=14)

# Display the plot
plt.show()
```
