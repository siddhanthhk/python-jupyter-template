### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Error handling
if len(data) == 0:
    raise ValueError("Data cannot be empty.")
if not all(isinstance(i, (int, float)) for i in data):
    raise ValueError("All elements in the data must be numeric.")

# Create the histogram
plt.hist(data, bins='auto', color='skyblue', edgecolor='black', rwidth=0.8)

# Add title and labels
plt.title("Histogram of Data", fontsize=14)
plt.xlabel("Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Optionally, add grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```
