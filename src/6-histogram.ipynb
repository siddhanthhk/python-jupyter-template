### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Error handling
if not data:
    raise ValueError("Data cannot be empty.")
if not all(isinstance(i, (int, float)) for i in data):
    raise ValueError("All elements in data must be numeric.")

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(
    data,
    bins='auto',
    color='skyblue',
    edgecolor='black',
    rwidth=0.85
)

# Add labels and title
plt.title("Histogram of Data", fontsize=14, fontweight='bold')
plt.xlabel("Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Grid for clarity
plt.grid(True, linestyle='--', alpha=0.6)

# Optional: Add vertical lines for mean/median
mean_val = sum(data) / len(data)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_val + 0.1, plt.ylim()[1]*0.9, f'Mean: {mean_val:.2f}', color='red')

# Show plot
plt.tight_layout()
plt.show()
```
