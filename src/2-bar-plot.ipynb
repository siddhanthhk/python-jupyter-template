### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # One color per bar

# Error handling
if len(categories) != len(values):
    raise ValueError("The 'categories' and 'values' lists must have the same length.")
if len(colors) != len(categories):
    raise ValueError("The 'colors' list must match the number of categories.")

# Create the plot
plt.figure(figsize=(8, 6), dpi=100)

# Create the bar plot
bars = plt.bar(categories, values, color=colors, edgecolor='black')

# Add a title and axis labels
plt.title("Bar Plot of Categories and Values", fontsize=16, fontweight='bold')
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Annotate bars with values
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.2, f'{yval}', 
             ha='center', va='bottom', fontsize=10)

# Optional: Add gridlines only on y-axis
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Optional: Save the plot
# plt.savefig("bar_plot_colored.png", dpi=300)

# Display the plot
plt.tight_layout()
plt.show()
```
