### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar

# Error handling
if len(categories) != len(values):
    raise ValueError("The 'categories' and 'values' lists must have the same length.")
if len(colors) != len(categories):
    raise ValueError("The 'colors' list must match the number of categories.")

# Create the bar plot
plt.bar(categories, values, color=colors)

# Add a title and labels
plt.title("Bar Plot of Categories and Values", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Annotate the bars with their values
for i, value in enumerate(values):
    plt.text(i, value + 0.2, str(value), ha='center', fontsize=10)

# Display the plot
plt.show()
```
