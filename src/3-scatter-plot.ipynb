### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error handling
if len(x) != len(y):
    raise ValueError("The 'x' and 'y' lists must have the same length.")
if len(x) == 0 or len(y) == 0:
    raise ValueError("The 'x' and 'y' lists must not be empty.")

# Create the scatter plot
plt.scatter(x, y, color='blue', label='Points')

# Annotate each point with its (x, y) value
for i, j in zip(x, y):
    plt.text(i, j, f'({i}, {j})', fontsize=9, ha='right', va='bottom')

# Add title and labels
plt.title("Scatter Plot with Annotations", fontsize=14)
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)

# Add a legend
plt.legend()

# Optionally, add a grid to make the plot clearer
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```
