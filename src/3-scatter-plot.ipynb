### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error handling
if len(x) != len(y):
    raise ValueError("The 'x' and 'y' lists must have the same length.")
if len(x) == 0 or len(y) == 0:
    raise ValueError("The 'x' and 'y' lists must not be empty.")

# Create the figure
plt.figure(figsize=(8, 6), dpi=100)

# Scatter plot
plt.scatter(x, y, color='blue', s=100, marker='o', edgecolor='black', label='Points')

# Annotate each point
for i, j in zip(x, y):
    plt.annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(-10, 10),
                 ha='center', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", fc="white", alpha=0.6))

# Titles and labels
plt.title("Scatter Plot with Annotations", fontsize=16, fontweight='bold')
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)

# Grid and legend
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

# Optional: Save the figure
# plt.savefig("scatter_plot_with_annotations.png", dpi=300)

# Display
plt.tight_layout()
plt.show()
```
