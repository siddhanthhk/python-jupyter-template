### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11, 1)

# Validate x
if not isinstance(x, np.ndarray) or x.size == 0:
    raise ValueError("x must be a non-empty numpy array.")

# Compute y = x^2
y = x**2

# Create the plot with specified figure size and DPI
plt.figure(figsize=(8, 6), dpi=100)

# Plot the line with markers
plt.plot(x, y, marker='o', linestyle='-', color='b', linewidth=2, label='y = x')

# Annotate each point with its coordinates
for i, j in zip(x, y):
    plt.text(i, j + 1, f'({i},{j})', ha='center', va='bottom', fontsize=9)

# Add title and axis labels
plt.title("Line Plot of y = x", fontsize=16, fontweight='bold')
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)

# Add grid, legend, and custom ticks
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(0, max(y)+10, 10))

# Optional: Save the figure
# plt.savefig("line_plot_x_squared.png", dpi=300)

# Show the plot
plt.tight_layout()
plt.show()
```
