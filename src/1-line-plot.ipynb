### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values ranging from 0 to 10
x = np.arange(0, 11, 1)

# Ensure x is a valid numeric array
if not isinstance(x, np.ndarray) or x.size == 0:
    raise ValueError("x must be a non-empty numpy array.")

# Calculate y values as the square of x
y = x**2

# Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

# Add a title and labels for both axes
plt.title("Line Plot of y = x²", fontsize=14)
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)

# Add a legend
plt.legend()

# Add a grid to the plot
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
